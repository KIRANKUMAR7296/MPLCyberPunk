# Importing Packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mplcyberpunk
plt.style.use('cyberpunk')
plt.rcParams['figure.figsize'] = (20,10)

font_title = {'Family' : 'Courier New', 'Size' : 24, 'Weight' : 'bold'}
font_label = {'Family' : 'Courier New', 'Size' : 20, 'Weight' : 'bold'}
```

# Importing Data Set

```python
data = pd.read_excel('Data/Canada.xlsx', sheet_name = 1, skiprows = range(20), skipfooter = 2)

data.head()
```

```python
data.info()
```

# Cleaning Data

**Drop** Unwanted Features.

```python
data.drop(['AREA','REG','DEV','Type','Coverage','DevName'], axis = 1, inplace = True)

data.head()
```

**Rename** Useful Features for Better Understanding.

```python
data.rename(columns = {'OdName' : 'Country',
                       'AreaName' : 'Continent',
                       'RegName' : 'Region'}, inplace = True)

data.head()                       
```

```python
data['Total'] = data.sum(axis = 1) 

data.head()
```

Setting Country as an **Index**

```python
data.set_index('Country', inplace = True)

data.tail()
```

```python
data.loc[['India','Pakistan','Sri Lanka','Nepal'],['Continent','Region','Total']]
```

Renaming Some Character Values with Large Names.

```python
data.rename(index = {'United Kingdom of Great Britain and Northern Ireland' : 'UK',
                     'United States of America' : 'USA',
                     'Iran (Islamic Republic of)' : 'Iran',
                     'Republic of Korea' : 'Korea'}, inplace = True) 

data.columns = data.columns.astype(str)

data.head()
```

```python
years = list(map(str, range(1980,2014)))

print(f'Years : \n{years}')
```

# 1.a Single Line Chart

A Seies of Data Points represented in a **Continuous Straight Line**. 

Example : Trends and Time Series Analysis. 

```python
print(f'India : \n{data.loc["India"].head()}\n')

print(f'Years Immigration \n{data.loc["India", years].head()}')
```

```python
data.loc['India', years].plot(kind = 'line')

# Title and Axes Labels :
plt.title('Immigration from India to Canada from 1980 to 2013', fontdict = font_title)
plt.xlabel('Years', fontdict = font_label)
plt.ylabel('Number of Immigrants', fontdict = font_label)

# Save Figure 
plt.savefig('Image/1.a.Single_Line_Chart.png');
```

# 1.b Multi Line Chart

Represents **Comparison** Trends.

```python
print(f'{"India".center(17,"-")}')
print(f'Years Immigration \n{data.loc["India", years].head()}\n')

print(f'{"Pakistan".center(17,"-")}')
print(f'Years Immigration \n{data.loc["Pakistan", years].head()}\n')

print(f'{"Sri Lanka".center(17,"-")}')
print(f'Years Immigration \n{data.loc["Sri Lanka", years].head()}')
```

```python
# Line Plots :
data.loc['India', years].plot(kind = 'line')
data.loc['Pakistan', years].plot(kind = 'line')
data.loc['Sri Lanka', years].plot(kind = 'line')

# Title and Axes Labels :
plt.title('Immigration from India, Pakistan and Sri Lanka to Canada from 1980 to 2013', fontdict = font_title)
plt.xlabel('Years', fontdict = font_label)
plt.ylabel('Number of Immigrants', fontdict = font_label)

# Legend :
plt.legend(fontsize = 16)

# Save Figure :
plt.savefig('Image/1.b.Multi_Line_Chart.png');
```

# 2.a Vertical Bar Chart

A Chart with **Rectangular** Bars with Length Proportional to the Values they Represent.

One Axis consist of **Categorical** Values.

```python
total = pd.DataFrame(data.loc[:, years].sum()) # Every Rows, Columns containing Only Years (1980 - 2013)

total.rename(columns = {0 : 'Total'}, inplace = True)

total.head()
```

```python
# Bar Plot : 
total.plot(kind = 'bar', legend = False)

# Title and Axes Labels :
plt.title('Total Immigrants to Canada from 1980 to 2013', fontdict = font_title)
plt.xlabel('Years', fontdict = font_label)
plt.ylabel('Number of Immigrants', fontdict = font_label)

# Save Figure :
plt.savefig('Image/2.a.Vertical_Bar_Chart.png');
```

# 2.b Horizontal Bar Chart 

Similar to Vertical Bar Chart just the Representation is **Horizontal**

```python
# Top 10 Countries with Highest Immigration :

top10 = pd.DataFrame(data.nlargest(10,'Total')['Total'].sort_values(ascending = True))

pd.DataFrame(data.nlargest(10,'Total')['Total'])
```

```python
# Bar Plot :

top10.plot.barh(color = 'crimson')

# Title and Axes Labels :
plt.title('Top 10 Immigrant Countries to Canada from 1980 to 2013', fontdict = font_title)
plt.xlabel('Number of Immigrants', fontdict = font_label)
plt.ylabel('Country', fontdict = font_label)

# Save Figure :
plt.savefig('Image/2.b.Horizontal_Bar_Chart.png');
```

# 2.c Vertical Grouped Bar Chart

Multiple Set of Data Items are **Compared**.

One Color is used to Denote a Specific **Series** across all **Sets**.

```python
year10 = list(map(str,(1980,1990,2000,2010)))

# Group of 4 Neighbouring Countries :
group = pd.DataFrame(data.loc[['India','Pakistan','Sri Lanka','Nepal'], year10].T)

group.head()
```

```python
# Bar Plot :
group.plot.bar(edgecolor = 'white')

# Title and Axes Labels :
plt.title('Total Immigrants to Canada from 1980 to 2010', fontdict = font_title)
plt.xticks(rotation = 0)
plt.xlabel('Year', fontdict = font_label)
plt.ylabel('Number of Immigrants', fontdict = font_label)

# Legend : 
plt.legend(fontsize = 16);

# Save Figure :
plt.savefig('Image/2.c.Vertical_Grouped_Bar_Chart.png');
```

# 2.d Horizontal Grouped Bar Chart

```python
# Bar Plot : 
group.plot.barh(edgecolor = 'white')

# Axes Title and Labels :
plt.title('Total Immigrants to Canada from 1980 to 2010', fontdict = font_title)
plt.xticks(rotation = 0)
plt.xlabel('Year', fontdict = font_label)
plt.ylabel('Number of Immigrants', fontdict = font_label)

# Legend : 
plt.legend(fontsize = 14);

# Save Figure : 
plt.savefig('Image/2.d.Horizontal_Grouped_Bar_Chart.png');
```

# 3.a Simple Area Chart

Depicts **Time Series** Relationship.

Unlike Line Plot, they Visually Represent **Volume**.

Information is Graphed on 2 Axes, using **Data Points** connected by **Line Segments**.

The **Area** between Axis and the Line is Emphasized with Color | Shading | Hatching.

```python
mplcyberpunk.add_glow_effects

plt.fill_between(total.index, total['Total'], color = 'greenyellow', alpha = 0.1)

plt.plot(total.index, total['Total'], color = 'palegreen', alpha = 1)

# Title and Axes Labels :
plt.title('Total Immigrants to Canada from 1980 to 2013', fontdict = font_title)
plt.xticks(rotation = 90)
plt.xlabel('Year', fontdict = font_title)
plt.ylabel('Number of Immigrants', fontdict = font_title)

# Save Figure :
plt.savefig('Image/3.a.Simple_Area_Chart.png');
```

# 3.b Stacked Area Chart

Displays the Value of Several Groups on Same Chart | On Top of Each Other.

```python
color = ['skyblue','greenyellow']

top = data.nlargest(2, 'Total').index.tolist()

top2 = pd.DataFrame(data.loc[top, years].T)

top2.plot(kind = 'area', stacked = True, color = color, alpha = 0.4)

# Title and Axes Labels :
plt.title('Total Immigrants to Canada from 1980 to 2013', fontdict = font_title)
plt.xlabel('Year', fontdict = font_label)
plt.ylabel('Number of Immigrants', fontdict = font_label)

# Legend :
plt.legend(fontsize = 14)

# Save Figure :
plt.savefig('Image/3.b.Stacked_Area_Chart.png');
```

# 3.c Unstacked Area Chart

Just Different Way of Arranging Data.

```python
# Area Plot :
top2.plot(kind = 'area', stacked = False, color = color, alpha = 0.4)

# Title and Axes Labels :
plt.title('Total Immigrants to Canada from 1980 to 2013', fontdict = font_title)
plt.xlabel('Year', fontdict = font_label)
plt.ylabel('Number of Immigrants', fontdict = font_label)

# Legend : 
plt.legend(fontsize = 14)

# Save Figure :
plt.savefig('Image/3.c.Unstacked_Area_Chart.png');
```

# 4.a Vertical Box Plot 

Also Known as **Whisker** and Box Plot.

**EDA** Exploratory Data Analysis | Statistical View of Data Frame.

Helps to Observe the Central Tendency and Dispersion i.e. **Skewness, Distribution** and **Outliers.**

```python
# Box Plot :
pd.DataFrame(data.loc[['India', 'Pakistan', 'Sri Lanka', 'USA', 'UK'],years].T).plot(kind = 'box')

# Title and Axes Labels :
plt.title('Immigrant Countries to Canada from 1980 to 2013', fontdict = font_title)
plt.xlabel('Country', fontdict = font_label)
plt.ylabel('Number of Immigrants', fontdict = font_label)

# Save Figure : 
plt.savefig('Image/4.a.Vertical_Box_Plot.png');
```

# 4.b Horizontal Box Plot

```python
# Box Plot :
pd.DataFrame(data.loc[['India', 'Pakistan', 'Sri Lanka', 'USA', 'UK'],years].T).plot(kind = 'box', vert = False)

# Title and Axes Labels :
plt.title('Immigrant Countries to Canada from 1980 to 2013', fontdict = font_title)
plt.xlabel('Number of Immigrants', fontdict = font_label)
plt.ylabel('Country', fontdict = font_label)

# Save Figure :
plt.savefig('Image/4.b.Horizontal_Box_Plot.png');
```

# 5. Scatter Plot

Displays Values for Two Variables using **Cartesian Coordinates**.

```python
iris = sns.load_dataset('iris')

iris.columns = [col.title() for col in iris.columns]

iris.head()
```

```python
# Scatter Plot :
sns.scatterplot('Sepal_Length', 'Sepal_Width', data = iris, s = 25, linewidth = 2, edgecolor = 'white')

# Title and Axes Labels :
plt.title('Scatter Plot', fontdict = font_title)
plt.xlabel('Sepal Length', fontdict = font_label)
plt.ylabel('Sepal Width', fontdict = font_label)

# Save Figure : 
plt.savefig('Image/5.Scatter.png');
```

# 6. Histogram

Groups Numbers into **Range**.

Frequency Distribution of given Variable | Numerical Data.

```python
# Histogram :
sns.distplot(iris['Sepal_Length'], color = 'lightpink', label = 'Sepal Length')
sns.distplot(iris['Sepal_Width'], color = 'lightblue', label = 'Sepal Width')

# Title and Axes Labels :
plt.title('Sepal Length and Width Frequency', fontdict = font_title)
plt.xlabel('Sepal Width', fontdict = font_label)
plt.ylabel('Frequency', fontdict = font_label)

# Legend :
plt.legend(fontsize = 14)

# Save Figure :
plt.savefig('Image/6.Histogram.png');
```

# 7. Bubble Plot

Similar to Scatter Plot with **3D** Appearance and Increased Size of Dots

```python
x,y = np.random.rand(1,30,1),np.random.rand(1,30,1)

size = np.random.rand(1,30,1)

# Scatter Plot :
plt.scatter(x,y,s = size * 1000, alpha = 0.4, color = 'red', edgecolor = 'white', linewidth = 2)

# Title :
plt.title('Bubble Plot', fontdict = font_title)

# Save Figure :
plt.savefig('Image/7.Bubble.png');
```

# 8. Pie Chart

Circular Statistical Graph, Divided into **Slices** of Numerical Proportion.


```python
disc = pd.DataFrame(data.groupby('Continent')['Total'].sum().T)

colors = ['gold','yellowgreen','lightcoral','lightskyblue','lightgreen','red']

explode  = [0,0.1,0,0,0.1,0.1]

# Pie Plot :
plt.pie(x = disc['Total'], explode = explode, colors = colors, startangle = 90, autopct = '%1.0f%%', pctdistance = 1.10, shadow = True)

# Title and Axes Labels :
plt.title('Continent Wise Immigrants Distribution', y = 1.1, fontdict = font_title)

plt.legend(disc.index, loc = 'upper left', fontsize = 14)

plt.axis('equal')

# Save Figure :
plt.savefig('Image/8.Pie.png');
```

# 9. Doughnut Chart

Variation of Pie Chart.

```python
top = data.nlargest(5,'Total').index.tolist()

top5 = pd.DataFrame(data.loc[top,'Total'].T)

top5.head()
```

```python
circle = plt.Circle((0,0),0.7, color = 'white')

colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']

text = {'color' : 'black', 'family' : 'Courier New', 'size' : 16, 'weight' : 'bold'}

# Pie Plot :
plt.pie(top5['Total'], explode = [0.1,0,0,0,0], colors = colors, autopct = '%1.0f%%', startangle = 90, shadow = True, textprops = text)

plt.gcf().gca().add_artist(circle)

plt.legend(top5.index, fontsize = 14, loc = 'upper left')

plt.axis('equal')

# Title :
plt.title('Top 5 Immigrant Country Distribution', fontdict = font_title)

# Save Figure :
plt.savefig('Image/9.Doughnut_Chart.png');
```

# 10.a Strong Trend Regression Plot

Exploratory Data Analysis **EDA**

Linear Relationships between Two **Quantitative** Variables

Positive Correlation.

```python
regression = pd.DataFrame(data.loc['India', years])

regression.reset_index(inplace = True)

regression.rename(columns = {'index' : 'Year'}, inplace = True)

regression[['Year','India']] = regression[['Year','India']].astype(int)

sns.regplot(x = 'Year', y = 'India', data = regression, color = 'b', scatter_kws = {'s' : 150, 'alpha' : 0.4})

# Title and Axes Labels :
plt.title('Total Immigration from India to Canada from 1980 to 2013', fontdict = font_title)
plt.xlabel('Year', fontdict = font_label)
plt.ylabel('Number of Immigrants', fontdict = font_label)

# Save Figure :
plt.savefig('Image/10.a.Strong_Regression_Line.png');
```

# 10.b Weak Trend Regression Plot

**Negative** Correlation.

```python
regression = pd.DataFrame(data.loc['Denmark', years])

regression.reset_index(inplace = True)

regression.rename(columns = {'index' : 'Year'}, inplace = True)

regression[['Year','Denmark']] = regression[['Year','Denmark']].astype(int)

sns.regplot(x = 'Year', y = 'Denmark', data = regression, color = 'r', scatter_kws = {'s' : 150, 'alpha' : 0.4})

# Title and Axes Labels :
plt.title('Total Immigration from Denmark to Canada from 1980 to 2013', fontdict = font_title)
plt.xlabel('Year', fontdict = font_label)
plt.ylabel('Number of Immigrants', fontdict = font_label)

# Save Figure : 
plt.savefig('Image/10.b.Weak_Regression_Line.png');
```

# 11. Word Cloud 

```python
plt.figure(figsize = (8,4))

from wordcloud import WordCloud, STOPWORDS

text = open(file = 'Data/Names.txt', mode = 'r', encoding = 'utf-8').read()

#stop = set(STOPWORDS)

# Word Cloud
cloud = WordCloud(background_color= 'white', max_words = 20)
cloud.generate(text)

plt.imshow(cloud, interpolation = 'bilinear')

plt.axis = ('off')

# Save Figure :
plt.savefig('Image/11.Word_Cloud.png');
```

# Removing Stopwords 

**Stop Words** are Words that are Very Frequently used and Add Absolutely no Meaning to the Text.

```python
plt.figure(figsize = (8,4))
stopwords = set(STOPWORDS)
stopwords.update(["stopword1", "stopword2"])

# Word Cloud
cloud = WordCloud(stopwords=stopwords, background_color='white').generate(text)
plt.imshow(cloud, interpolation="bilinear")
plt.tight_layout(pad=0);
```
